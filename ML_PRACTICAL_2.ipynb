### Mean Squared Error (MSE)
is a commonly used metric to evaluate the performance of a regression model. It measures the average squared difference between the predicted values and the actual values. [ 1/2(yi-yi')^2 ]

### Root Mean Squared Error (RMSE)
is the square root of the Mean Squared Error (MSE). It provides a measure of the average magnitude of the errors in the predictions, but unlike MSE, it is in the same units as the original data, making it easier to interpret. (under root MSE)

### R² (R-squared)
is a measure of how well the regression model's predictions match the actual data. (1- summation(yi-yi')^2 / summation(yi-yi')^2)

**MSE** is useful when you want to penalize large errors more severely.e.

**RMSE**  is preferred when you want a more interpretable measure that is in the same units as the original data.

**R²** is useful for understanding how well the model fits the data in a relative sense.

**Lasso Regression (L1 Regularization)**
Lasso regression adds a penalty to the loss function based on the sum of the absolute values of the coefficients.

**Effect:** Lasso can shrink some coefficients exactly to zero, effectively performing feature selection and eliminating irrelevant variables.
.

**Ridge Regression (L2 Regularization)**
Ridge regression adds a penalty to the loss function based on the sum of the squared values of the coefficients.

**Effect:** It shrinks the coefficients but does not set them exactly to zero, so it is used when you want to retain all features but with reduced impact.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df=pd.read_csv("uber.csv")
```

```python
df.dropna(inplace=True)
```

```python
df = df.drop(['Unnamed: 0', 'key'], axis=1)
```

```python
df = df[(df['fare_amount'] > 0) & (df['passenger_count'] > 0)]
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

df['hour'] = df['pickup_datetime'].dt.hour

df = df.drop(['pickup_datetime'], axis=1)

df["price"] = df['hour']*df['fare_amount']
```

```python
df.head()
```

```python
#outliers
Q1 = df['fare_amount'].quantile(0.25)
Q3 = df['fare_amount'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['fare_amount'] >= lower_bound) & (df['fare_amount'] <= upper_bound)]
```

```python
corr_matrix=df.corr()
corr_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title('Correlation Matrix')
plt.show()
```

```python
x = df.drop(['fare_amount'], axis=1)
y = df['fare_amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_test)
```

```python
ridge_model = Ridge()

ridge_model.fit(X_train, y_train)

y_pred_ridge = ridge_model.predict(X_test)
```

```python

lasso_model = Lasso()

lasso_model.fit(X_train, y_train)

y_pred_lasso = lasso_model.predict(X_test)
```

```python
# from sklearn.metrics import r2_score, mean_squared_error

def evaluate_model(y_test, y_pred, model_name):
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f'{model_name} - R2: {r2:.4f}, RMSE: {rmse:.2f}')
```

```python

evaluate_model(y_test, y_pred_lr, 'Linear Regression')
evaluate_model(y_test, y_pred_ridge, 'Ridge Regression')
evaluate_model(y_test, y_pred_lasso, 'Lasso Regression')
```
